```python
def make_env():
    def _init():
        return FlexibleTradingEnv(
            df=df,
            prices=prices,
            signal_features=signal_features,
            window_size=window_size,
            frame_bound=frame_bound,
            fee=0.0005,
            holding_cost=0.00001,
            render_mode=None
        )
    return _init
```

```python
import pickle
from stable_baselines3.common.vec_env import DummyVecEnv, VecNormalize, SubprocVecEnv

train_env = SubprocVecEnv([make_env() for _ in range(4)])
train_env = VecNormalize(train_env, norm_obs=True, norm_reward=True, clip_reward=np.inf)

eval_env = DummyVecEnv([make_env()])  # only 1 env

# Try to load normalization safely and print shapes for debugging
try:
    # prefer using VecNormalize.load when shapes match
    saved = VecNormalize.load("vec_normalize.pkl", eval_env)
    eval_env = saved
    print("Loaded VecNormalize directly (shapes matched).")
except Exception as e:
    print("VecNormalize.load failed:", e)
    # fallback: open pickle, inspect shapes, then copy stats if compatible
    with open("vec_normalize.pkl", "rb") as f:
        saved = pickle.load(f)
    print("Saved ob_rms mean shape:", getattr(saved.ob_rms, "mean", None).shape if hasattr(saved, "ob_rms") else None)
    print("Saved ret_rms shape:", getattr(saved, "ret_rms", "N/A"))
    # wrap eval_env and copy RMS if obs dims match
    eval_env = VecNormalize(eval_env, norm_obs=True, norm_reward=False, clip_reward=np.inf)
    try:
        eval_env.ob_rms = saved.ob_rms
        eval_env.ret_rms = saved.ret_rms
        print("Copied ob_rms and ret_rms to eval_env.")
    except Exception as copy_err:
        print("Failed to copy RMS stats:", copy_err)
        # print shapes to help debug
        print("eval env observation_space:", eval_env.observation_space)
        raise

eval_env.training = False
eval_env.norm_reward = False
```
