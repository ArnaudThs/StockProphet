```python
import sqlite3
import pandas as pd
```

```python
db_path = "AAPL_stock_news.db"
conn = sqlite3.connect(db_path)
conn
```

```python
df = pd.read_sql_query("SELECT * FROM news_articles", conn)
df.head()
```

```python
df
```

```python
import pandas as pd

# Create cleaned dataframe with only the needed columns
df_clean = df[['published_utc', 'title', 'description']].copy()

# Convert published_utc to date (day only) and rename to "date"
df_clean['date'] = pd.to_datetime(df_clean['published_utc']).dt.date
df_clean = df_clean.drop(columns=['published_utc'])

# Optional: reorder columns
df_clean = df_clean[['date', 'title', 'description']]
df_clean.head()
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
```

```python
MODEL_NAME = "yiyanghkust/finbert-tone"

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForSequenceClassification.from_pretrained(MODEL_NAME)

finbert = pipeline(
    "sentiment-analysis",
    model=model,
    tokenizer=tokenizer,
    truncation=True,
    max_length=256,   # évite les textes trop longs
    # device=0        # décommente si tu as un GPU
)
```

```python
import pandas as pd

# Sécurité: remplacer les NaN
df_clean["title"] = df_clean["title"].fillna("")
df_clean["description"] = df_clean["description"].fillna("")

# Texte à analyser = titre + description
df_clean["text_for_sentiment"] = (df_clean["title"] + ". " + df_clean["description"]).str.strip()

# On garde seulement les lignes qui ont du texte
mask = df_clean["text_for_sentiment"] != ""
subset = df_clean.loc[mask].copy()

len(subset), subset.head(2)
```

```python
batch_size = 32

sentiments = []
scores = []

for i in range(0, len(subset), batch_size):
    batch_texts = subset["text_for_sentiment"].iloc[i:i+batch_size].tolist()
    results = finbert(batch_texts)

    for res in results:
        label = res["label"]  # ex: 'positive', 'negative', 'neutral' (ou parfois 'Positive')
        score = float(res["score"])

        # Normaliser le label
        norm_label = label.lower().strip()
        if norm_label.startswith("pos"):
            norm_label = "positive"
        elif norm_label.startswith("neg"):
            norm_label = "negative"
        elif norm_label.startswith("neu"):
            norm_label = "neutral"

        sentiments.append(norm_label)
        scores.append(score)

len(sentiments), len(scores)
```

```python
subset.loc[:, "finbert_sentiment"] = sentiments
subset.loc[:, "finbert_score"] = scores

subset[["date", "title", "finbert_sentiment", "finbert_score"]].head()
```

```python
# On joint les colonnes de subset dans df_clean (en respectant les index)
df_clean = df_clean.join(
    subset[["finbert_sentiment", "finbert_score"]],
    how="left"
)

df_clean[["date", "title", "finbert_sentiment", "finbert_score"]].head()
```

## Here we have the database with finbert sentiment & score

```python
df_clean.tail(10)
```

```python
sentiment_map = {
    "positive": 1,
    "neutral": 0,
    "negative": -1
}

df_clean["sentiment_num"] = df_clean["finbert_sentiment"].map(sentiment_map)
```

```python
df_daily = (
    df_clean
    .groupby("date", as_index=False)["sentiment_num"]
    .sum()
)

# Renommer pour que ce soit plus clair
df_daily = df_daily.rename(columns={"sentiment_num": "daily_sentiment"})
df_daily.tail(30)
```

```python
import sqlite3

DB_PATH = "AAPL_daily_sentiment.db"  # nouveau fichier .db

conn = sqlite3.connect(DB_PATH)

# On écrit df_daily dans une table "daily_sentiment"
df_daily.to_sql("daily_sentiment", conn, if_exists="replace", index=False)

conn.close()

print("✅ Base SQLite créée :", DB_PATH)
```
