```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("stock_news.db")

df_aapl = pd.read_sql_query("""
    SELECT published_utc, finbert_sentiment
    FROM news_articles
    WHERE ticker = 'AAPL'
""", conn)

conn.close()

df_aapl.head()
```

```python
df_aapl
```

```python
df_aapl["published_utc"] = pd.to_datetime(df_aapl["published_utc"])
```

```python
df_aapl["date"] = df_aapl["published_utc"].dt.date
```

```python
df_aapl = df_aapl[["date", "finbert_sentiment"]]
df_aapl.head()
```

```python
mapping = {"positive": 1, "neutral": 0, "negative": -1}

df_aapl["sentiment_score"] = df_aapl["finbert_sentiment"].map(mapping)
df_aapl.head()
```

```python
df_aapl = df_aapl.drop(columns=["finbert_sentiment"])
```

```python
df_aapl = df_aapl.groupby("date")["sentiment_score"].sum().reset_index()
```
